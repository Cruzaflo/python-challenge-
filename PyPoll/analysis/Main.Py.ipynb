```python
import os
import csv

voter_id = []

county = []

candidate_list = []

cand_votes = {}

total_votes = 0 

majority = 0

#Files
csvpath = os.path.join("..", "Resources", "election_data.csv")
output_file = os.path.join("election_results.txt")

#output_file = csvpath = os.path.join("final_results.txt")

#ID, County, Candidate
with open(csvpath, newline = "", encoding='utf-8') as csvfile:
    datafile = csv.reader(csvfile, delimiter=",")
    next(datafile, None)
    
    for row in datafile:
        
        total_votes += 1
        
        candidate = row[2]
        
        if candidate not in candidate_list:
            
            candidate_list.append(candidate)
            
            cand_votes[candidate] = 0 

        cand_votes[candidate] += 1
    
with open(output_file, 'w', encoding='utf-8') as txt_file:
    header = (
        "Election Results\n" +
        "-------------------------\n"+
        "Total Votes: " + str(total_votes) + "\n" +
        "-------------------------\n"
    )  
    print(header, end="")
    txt_file.write(header)
    
    for name in cand_votes:
        #Candidate vote percentage
        votes = cand_votes.get(name)
        percentage = float(votes) / float(total_votes) * 100 

        #The winning candidate
        if votes > majority:
            majority = votes
            winner = name
        
        #Output results
        output = f"{name}: {percentage:.3f}% ({votes})\n"
        
        #print and write to file
        print(output)
        txt_file.write(output)
        
    #printer and write winner to file
    winner_output = (
        "-------------------------\n" +
        "Winner: " + (winner) + "\n" +
        "-------------------------\n"
    )
    
    print(winner_output)
    
    txt_file.write(winner_output)
    
    
```
